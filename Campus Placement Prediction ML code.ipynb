```python
df1=df.copy()
df1.drop(['Email','Name'], axis=1, inplace=True)
```

```python
df1
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
lst=['Gender','10th board','12th board','Stream', 'Internships(Y/N)', 
     'Training(Y/N)', 'Backlog in 5th sem', 'Project(Y/N)', 
     'Communication level', 'Technical Course(Y/N)', 'Placement(Y/N)?']
for i in lst:
    df1[i]=le.fit_transform(df1[i])
    
df1
```

```python
df1.dtypes
```

```python
X=df1.iloc[:,:13]
y=df1.iloc[:,-1]
```

```python
print(X)
```

```python
#scaling the dataset using Min-Max scaler
from sklearn.preprocessing import MinMaxScaler
ms=MinMaxScaler()
X_sc=ms.fit_transform(X)
X_sc
```

```python
#splitting the dataset to train data and test data
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X_sc,y,test_size=0.2,random_state=1)
```

## AdaBoost Classifier

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.metrics import ConfusionMatrixDisplay,classification_report

ad=AdaBoostClassifier()
ad.fit(X_train,y_train)
y_pred1=ad.predict(X_test)
print(classification_report(y_test,y_pred1))
```

## Logistic Regression Classifier

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(random_state=0, solver='lbfgs', max_iter=1000).fit(X_train, y_train)
# printing the acc
y_pred2=clf.predict(X_test)
print(classification_report(y_test,y_pred2))
```

## Support Vector Classifier

```python
from sklearn.svm import SVC

svc=SVC()
svc.fit(X_train,y_train)
y_pred3=svc.predict(X_test)
print(classification_report(y_test,y_pred3))
```

## Naive Bayes Classifier

```python
from sklearn.naive_bayes import GaussianNB

nb=GaussianNB()
nb.fit(X_train,y_train)
y_pred4=nb.predict(X_test)
print(classification_report(y_test,y_pred4))
```

## Gradient Boosting Classifier

```python
from sklearn.ensemble import GradientBoostingClassifier

gb=GradientBoostingClassifier()
gb.fit(X_train,y_train)
y_pred5=gb.predict(X_test)
print(classification_report(y_test,y_pred5))
```
